### Categorical features

```python
categorical_cols = ['traffic_control_device', 'device_condition', 'weather_condition', 
                    'lighting_condition', 'first_crash_type', 'trafficway_type', 
                    'alignment', 'roadway_surface_condition', 'road_defect', 
                    'crash_type']

for col in categorical_cols:
    data[col] = data[col].str.upper().str.strip()
```

```python
# Let's find the categorical columns
categorical_columns = data.select_dtypes(include=['object']).columns.tolist()
print(categorical_columns)
```

```python
# unique values in each categorical column
for column in categorical_columns:
    print(f'{column} unique values: {data[column].unique().tolist()}')
```

### Numerical features

```python
invalid_speeds = data[(data['posted_speed_limit'] < 0) | (data['posted_speed_limit'] > 100)]
print(invalid_speeds)

data = data[(data['posted_speed_limit'] >= 0) & (data['posted_speed_limit'] <= 100)]
```

```python
sns.countplot(x='crash_type', data=data)
plt.title('Crash Type Distribution')
plt.show()

# Calculate class imbalance
class_counts = data['crash_type'].value_counts()
print(class_counts)
```

```python
from sklearn.utils import resample

# Separate majority and minority classes
majority = data[data.crash_type == 'NO INJURY / DRIVE AWAY']
minority = data[data.crash_type == 'INJURY AND / OR TOW DUE TO CRASH']

# Downsample majority class
majority_downsampled = resample(majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=len(minority),  # to match minority class
                                 random_state=42) 

# Combine minority class with downsampled majority class
data_balanced = pd.concat([majority_downsampled, minority])
```

```python
sns.countplot(x='crash_type', data=data_balanced)
plt.title('Crash Type Distribution')
plt.show()

# Calculate class imbalance
class_counts_balanced = data_balanced['crash_type'].value_counts()
print(class_counts_balanced)
```

### Feature Importance

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
X = data.drop('crash_type', axis=1)
y = data['crash_type']
```

```python
le = LabelEncoder()
X_encoded = X.apply(lambda col: le.fit_transform(col.astype(str)) if col.dtype == 'object' else col)
```

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_encoded, y)
```

```python
importances = rf.feature_importances_
feature_names = X.columns
feature_importances = pd.Series(importances, index=feature_names).sort_values(ascending=False)
```

```python
plt.figure()
sns.barplot(x=feature_importances[:20], y=feature_importances.index[:20])
plt.title('Top 20 Feature Importances')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()
```

```python
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns
corr_matrix = data[numerical_features].corr()

plt.figure()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Let's drop the features that have less than 0.01 importance score
low_importance_features = feature_importances[feature_importances < 0.01].index
X.drop(columns=low_importance_features, inplace=True)
print(low_importance_features)
```
